```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
coaches = pd.read_csv('coaches.csv')
coaches
```

```python
teams= pd.read_csv('teams.csv')
teams
```

```python
teams_coaches = pd.merge(teams, coaches, on='coach_id', suffixes=('_team', '_coach'))
teams_coaches
```

```python
leagues = pd.read_csv('leagues.csv')
teams_coaches_leagues = pd.merge(teams_coaches, leagues, on='league_id', suffixes=('_team', '_league'))
teams_coaches_leagues
```

```python
stadiums=pd.read_csv('stadiums.csv')
teams_coaches_leagues_stadiums = pd.merge(teams_coaches_leagues, stadiums, on='stadium_id' )
teams_coaches_leagues_stadiums
```

```python
teams_coaches_leagues_stadiums.rename(columns={'name_x': 'league_name', 'name_y': 'stadium_name'}, inplace=True)
teams_coaches_leagues_stadiums
```

```python
teams_coaches_leagues_stadiums.drop(columns=['cresturl', 'location', 'capacity'], inplace=True)
teams_coaches_leagues_stadiums
```

```python
teams_coaches_leagues_stadiums.rename(columns={'team_id_team': 'team_id'}, inplace=True)
df = teams_coaches_leagues_stadiums.copy()
players = pd.read_csv('players.csv')
```

```python
df_players = pd.merge(df, players, on='team_id', how='left')
df_players
```

```python
df_players.columns
```

```python

df_players.rename(columns={'nationality_x': 'nationality_coach', 'nationality_y': 'nationality_player'}, inplace=True)
df_players
```

```python
standings = pd.read_csv('standings.csv')
standings=standings[['goals_for', 'goals_against', 'goal_difference', 'points', 'team_id']]
df_players_standings = pd.merge(df_players, standings, on='team_id', how='left')
df_players_standings
```
